```python
import random
print("*******Welcome to RPS*******")
user_score =0
comp_score =0
tie = 0
player_name = input("Enter your name :")
print("""Winning Rules are
1.Paper vs Rock ---> Paper
2.Rock vs Scissors ---> Rock
3.Scissor vs Paper ---> Scissor""")
print ("-------------------------------------------------")
while True:
    print("""Select the number 
    1.Rock
    2.Paper
    3.Scissor """)
    print ("-------------------------------------------------")
    choice = int(input("Enter the number from 1 -3 :"))
    print ("-------------------------------------------------")
    while choice > 3 or choice < 1:
        choice = int(input("Enter the valid number :"))

    if choice ==1:
        user_choice = "Rock"
    elif choice ==2:
        user_choice ="Paper"
    else:
        user_choice = "Scissor"
    print("the user choice is :" ,user_choice)
    print ("-------------------------------------------------")


    computer = random.randint(1,3)
    if computer ==1:
        comp_choice = "Rock"
    elif computer ==2:
        comp_choice = "Paper"
    else :
        comp_choice = "Scissor"
        print ("-------------------------------------------------")
    print("Computer choice is : ",comp_choice )
    print ("-------------------------------------------------")
    if ((user_choice == "Paper" and comp_choice =="Rock") or (comp_choice == "Paper" and user_choice == "Rock")):
        print("**Paper Wins**")
        result = "Paper"
    elif((user_choice == "Rock" and comp_choice =="Scissor") or (comp_choice == "Scissor" and user_choice == "Rock")):
        print("**Rock Wins**")
        result = "Rock"
    elif(user_choice == comp_choice):
        print("it is tie")
        result = "tie"
    else:
        print("**Scissor Wins**")
        result = "Scissor"
    
    if result == "tie":
         tie +=1
    elif result == user_choice:
        print("**User Wins**")
        user_score +=1
    else:
        comp_score +=1
    print ("-------------------------------------------------")
    print("Scores are : ")
    print(player_name , "Score is :" , user_score)
    print("computer score is :", comp_score)
    print("ties are :" , tie)
    repeat = input ("do you want to play again?")
    if repeat =="No" :
        break
print("Game Over")
```
